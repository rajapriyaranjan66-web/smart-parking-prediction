```python
import pandas as pd

data = pd.read_csv("parking_data.csv")
print(data.head())
```

```python
# Convert date_time to proper format
data['date_time'] = pd.to_datetime(data['date_time'])

# Extract useful features
data['hour'] = data['date_time'].dt.hour
data['day_of_week'] = data['date_time'].dt.dayofweek

# Create target (0 = Available, 1 = Full)
data['target'] = data['occupancy_rate_percent'].apply(lambda x: 1 if x > 80 else 0)

print(data[['hour','day_of_week','target']].head())
```

```python
data = pd.get_dummies(data, columns=['zone','vehicle_type'], drop_first=True)
```

```python
X = data[['hour','day_of_week','entry_count','exit_count']]
y = data['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
import pandas as pd

data = pd.read_csv("parking_data.csv")
data.head()
```

```python
data['date_time'] = pd.to_datetime(data['date_time'])

data['hour'] = data['date_time'].dt.hour
data['day_of_week'] = data['date_time'].dt.dayofweek

data['target'] = data['occupancy_rate_percent'].apply(lambda x: 1 if x > 80 else 0)

data = pd.get_dummies(data, columns=['zone','vehicle_type'], drop_first=True)

X = data[['hour','day_of_week','entry_count','exit_count']]
y = data['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```
